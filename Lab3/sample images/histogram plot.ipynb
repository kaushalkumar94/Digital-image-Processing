```python
import cv2
import matplotlib.pyplot as plt

# ===== IMAGE PATHS =====
image_paths = [
    r"D:\i bet you open this\WALL\3. BATMAN.jpg"
]

# ===== CREATE FIGURE =====
plt.figure(figsize=(14, 8))

for i, path in enumerate(image_paths):

    # Read image in grayscale
    img = cv2.imread(path, 0)

    if img is None:
        print(f"Image not found: {path}")
        continue

    # ---- Show Image ----
    plt.subplot(4, 2, 2*i + 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"Image {i+1}")
    plt.axis('off')

    # ---- Show Histogram ----
    plt.subplot(4, 2, 2*i + 2)
    plt.hist(img.ravel(), bins=256, range=[0,256], color='black')
    plt.title(f"Histogram {i+1}")
    plt.xlabel("Pixel Intensity")
    plt.ylabel("Frequency")

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create 16x16 black image
img = np.zeros((16,16), dtype=np.uint8)

# Divide horizontally
img[0:8, :] = 255   # Top half = white
img[8:16, :] = 0    # Bottom half = black

# Display image
plt.figure(figsize=(4,4))
plt.imshow(img, cmap='gray', interpolation='nearest')
plt.title("16x16 Image - Horizontal Division")
plt.axis('off')
plt.show()
```

```python
import numpy as np
import cv2

# folder path to save images
save_path = r"D:\CODING\DIP\LAB3\sample images"

# create checkerboard image (16x16 with 8x8 blocks)
checker = np.zeros((16,16), dtype=np.uint8)
checker[0:8, 0:8] = 255
checker[8:16, 8:16] = 255
cv2.imwrite(save_path + r"\checkerboard_16x16.png", checker)

# create horizontal division image
horizontal = np.zeros((16,16), dtype=np.uint8)
horizontal[0:8, :] = 255
cv2.imwrite(save_path + r"\horizontal_16x16.png", horizontal)

# create vertical division image
vertical = np.zeros((16,16), dtype=np.uint8)
vertical[:, 0:8] = 255
cv2.imwrite(save_path + r"\vertical_16x16.png", vertical)

print("Images saved")
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# custom function to calculate histogram without ravel()
def custom_histogram(image_array):
    hist = np.zeros(256, dtype=int)

    rows, cols = image_array.shape

    # loop through every pixel
    for i in range(rows):
        for j in range(cols):
            pixel_value = image_array[i][j]
            hist[pixel_value] += 1

    return hist


# load image in grayscale
img = Image.open(r"D:\i bet you open this\WALL\3. BATMAN.jpg").convert("L")

# convert image to numpy array
img_array = np.array(img)

# get histogram using custom function
hist = custom_histogram(img_array)

# plot histogram
plt.figure(figsize=(8,5))
plt.plot(hist, color='black')
plt.title("Custom Histogram")
plt.xlabel("Pixel Intensity (0-255)")
plt.ylabel("Number of Pixels")
plt.show()
```
